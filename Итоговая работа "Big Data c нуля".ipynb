# **1. Загрузите в колаб файлы по оценкам (ratings) и фильмам (movies) и создайте на их основе pandas-датафреймы**

```python
import pandas as pd
```

```python
columns = ['User_Id', 'Item_id', 'Rating', 'Date']
ratings = pd.read_csv('u.data.csv', sep='\s+', header = None, names = columns)
ratings.head()
```

```python
columns = ['Movie_id', 'movie_title', 'release_date', 'video_release_date', 'IMDb_URL', 'unknown', 'Action', 'Adventure', 'Animation', 'Children', 'Comedy',
           'Crime', 'Documentary', 'Drama', 'Fantasy', 'Film_Noir', 'Horror', 'Musical', 'Mystery', 'Romance', 'Sci_Fi',
           'Thriller','War','Western']
movies = pd.read_csv('u.item.csv', sep='|', encoding='latin-1', header = None, names = columns)
movies.head()
```

# **2. Средствами Pandas, используя dataframe ratings, найдите id пользователя, поставившего больше всего оценок**

```python
ratings[['User_Id','Item_id']].groupby('User_Id').count().sort_values('Item_id', ascending=False).head(1)
```

# **3. Оставьте в датафрейме ratings только те фильмы, который оценил данный пользователь** 

```python
rate_405 = ratings[ratings['User_Id'] == '405']
rate_405
```

```python
ratings_405 = ratings[(ratings['User_Id'] > 404) & (ratings['User_Id'] < 406)]
ratings_405.head(5)
```

# **4. Добавьте к датафрейму из задания 3 столбцы:**
1.   По жанрам. Каждый столбец - это жанр. Единицу записываем, если фильм принадлежит данному жанру и 0 - если нет
2.   Столбцы с общим количеством оценок от всех пользователей на фильм и суммарной оценкой от всех пользователей

Присоединяем столбцы жанра

```python
RM_405 = ratings_405.merge(movies, left_on='Item_id', right_on='Movie_id')
RM_405.head(5)
```

Вычисляем средний рейтинг на основе всех рейтингов

```python
score = ratings[['Item_id','Rating']].groupby('Item_id').mean()
score.head(5)
```

Вычисляем количество голосов

```python
count_r = ratings[['Item_id','Rating']].groupby('Item_id').count()
count_r.head(5)
```

Объединяем средний рейтинг и количество голосов

```python
rating_all = count_r.merge(score, left_on='Item_id', right_on='Item_id')
rating_all.head(5)
```

Присоединяем получившиеся значения к общей таблице

```python
RMR_405 = RM_405.merge(rating_all, left_on='Item_id', right_on='Item_id')
RMR_405.head(5)
```

Убираем ненужные столбцы

```python
RMR_405.columns.tolist()
```

```python
RMR_405 = RMR_405[['User_Id', 'Movie_id', 'movie_title', 'release_date', 'Rating','Rating_x', 'Rating_y', 'unknown',
 'Action', 'Adventure', 'Animation', 'Children', 'Comedy', 'Crime', 'Documentary', 'Drama', 'Fantasy', 'Film_Noir',
 'Horror', 'Musical', 'Mystery', 'Romance', 'Sci_Fi', 'Thriller', 'War', 'Western']]
```

```python
RMR_405.head()
```

# Построение модели
**Сформируйте X_train, X_test, y_train, y_test**

```python
X = RMR_405[['Rating_x', 'Rating_y', 'unknown', 'Action', 'Adventure', 'Animation', 'Children', 'Comedy', 'Crime',
             'Documentary', 'Drama', 'Fantasy', 'Film_Noir', 'Horror', 'Musical', 'Mystery', 'Romance', 'Sci_Fi', 'Thriller', 'War', 'Western']]
Y = RMR_405['Rating']
```

**Возьмите модель линейной регрессии (или любую другую для задачи регрессии)  и обучите ее на фильмах**

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.1)
```

```python
lr = LinearRegression()
lr.fit(X_train, Y_train)
```

**Оцените качество модели на X_test, y_test при помощи метрик для задачи регрессии**

```python
mean_squared_error(Y_test, lr.predict(X_test)).T
```

```python
test = PYratings.withColumn('unknown', when(col('unknown').isin('1'), "unknown")\
                            .otherwise(when(col('unknown').isin('0'), "")
                            .otherwise(PYratings.unknown)))
test = test.withColumn('Action', when(col('Action').isin('1'), "Action")\
                            .otherwise(when(col('Action').isin('0'), "")
                            .otherwise(test.Action)))
test = test.withColumn('Adventure', when(col('Adventure').isin('1'), "Adventure")\
                            .otherwise(when(col('Adventure').isin('0'), "")
                            .otherwise(test.Adventure)))
test = test.withColumn('Animation', when(col('Animation').isin('1'), "Animation")\
                            .otherwise(when(col('Animation').isin('0'), "")
                            .otherwise(test.Animation)))
test = test.withColumn('Children', when(col('Children').isin('1'), "Children")\
                            .otherwise(when(col('Children').isin('0'), "")
                            .otherwise(test.Children)))
test = test.withColumn('Comedy', when(col('Comedy').isin('1'), "Comedy")\
                            .otherwise(when(col('Comedy').isin('0'), "")
                            .otherwise(test.Comedy)))
test = test.withColumn('Crime', when(col('Crime').isin('1'), "Crime")\
                            .otherwise(when(col('Crime').isin('0'), "")
                            .otherwise(test.Crime)))
test = test.withColumn('Documentary', when(col('Documentary').isin('1'), "Documentary")\
                            .otherwise(when(col('Documentary').isin('0'), "")
                            .otherwise(test.Documentary)))
test = test.withColumn('Drama', when(col('Drama').isin('1'), "Drama")\
                            .otherwise(when(col('Drama').isin('0'), "")
                            .otherwise(test.Drama)))
test = test.withColumn('Fantasy', when(col('Fantasy').isin('1'), "Fantasy")\
                            .otherwise(when(col('Fantasy').isin('0'), "")
                            .otherwise(test.Fantasy)))
test = test.withColumn('Film_Noir', when(col('Film_Noir').isin('1'), "Film_Noir")\
                            .otherwise(when(col('Film_Noir').isin('0'), "")
                            .otherwise(test.Film_Noir)))
test = test.withColumn('Horror', when(col('Horror').isin('1'), "Horror")\
                            .otherwise(when(col('Horror').isin('0'), "")
                            .otherwise(test.Horror)))
test = test.withColumn('Musical', when(col('Musical').isin('1'), "Musical")\
                            .otherwise(when(col('Musical').isin('0'), "")
                            .otherwise(test.Musical)))
test = test.withColumn('Mystery', when(col('Mystery').isin('1'), "Mystery")\
                            .otherwise(when(col('Mystery').isin('0'), "")
                            .otherwise(test.Mystery)))
test = test.withColumn('Romance', when(col('Romance').isin('1'), "Romance")\
                            .otherwise(when(col('Romance').isin('0'), "")
                            .otherwise(test.Romance)))
test = test.withColumn('Sci_Fi', when(col('Sci_Fi').isin('1'), "Sci_Fi")\
                            .otherwise(when(col('Sci_Fi').isin('0'), "")
                            .otherwise(test.Sci_Fi)))
test = test.withColumn('Thriller', when(col('Thriller').isin('1'), "Thriller")\
                            .otherwise(when(col('Thriller').isin('0'), "")
                            .otherwise(test.Thriller)))
test = test.withColumn('War', when(col('War').isin('1'), "War")\
                            .otherwise(when(col('War').isin('0'), "")
                            .otherwise(test.War)))
test = test.withColumn('Western', when(col('Western').isin('1'), "Western")\
                            .otherwise(when(col('Western').isin('0'), "")
                            .otherwise(test.Western)))
```
