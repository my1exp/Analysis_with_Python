```python
import pandas as pd
btc_price = pd.read_csv('BTC-USD.csv', ',')
ltc_price = pd.read_csv('LTC-USD.csv', ',')
btc_adress = pd.read_json('total-addresses-btc-24h.json')
ltc_adress = pd.read_json('total-addresses-ltc-24h.json')
btc_fees = pd.read_json('fees-mean-btc-24h.json')
ltc_fees = pd.read_json('fees-mean-ltc-24h.json')
btc_tx_count = pd.read_json('transaction-count-btc-24h.json')
ltc_tx_count = pd.read_json('transaction-count-ltc-24h.json')
```

```python
df_full = btc_price.merge(ltc_price, left_on='Date', right_on='Date')
df_other = btc_adress.merge(ltc_adress, left_on='t', right_on='t')
df_other.rename(columns = {'v_x' : 'Btc adress', 'v_y' : 'Ltc adress'}, inplace = True)
df_other = df_other.merge(btc_fees, left_on='t', right_on='t')
df_other = df_other.merge(ltc_fees, left_on='t', right_on='t')
df_other.rename(columns = {'v_x' : 'Btc fees', 'v_y' : 'Ltc fees'}, inplace = True)
df_other = df_other.merge(btc_tx_count, left_on='t', right_on='t')
df_other = df_other.merge(ltc_tx_count, left_on='t', right_on='t')
df_other.rename(columns = {'v_x' : "Btc tx's", 'v_y' : "Ltc tx's"}, inplace = True)
df_other = df_other[(df_other['t'] > '2014-09-17')]
df_other.reset_index(drop=True, inplace=True)
df_full = df_full[(df_full['Date'] < '2022-02-28')]
```

```python
df_full = pd.DataFrame({'Date': pd.date_range('2014-09-17', periods= len(df_full), freq='1D'),
                   'Btc open' : df_full['Open_x'],
                   'Btc close' : df_full['Close_x'],
                   'Btc volume' : df_full['Volume_x'],
                   'Btc adress' : df_other['Btc adress'],
                   "Btc tx's" : df_other["Btc tx's"],
                   'Btc fees' : df_other['Btc fees'],
                   'Ltc open' : df_full['Open_y'],
                   'Ltc close' : df_full['Close_y'],
                   'Ltc volume' : df_full['Volume_y'], 
                   'Ltc adress' : df_other['Ltc adress'],
                   "Ltc tx's" : df_other["Ltc tx's"],
                   'Ltc fees' : df_other['Ltc fees'],    
                        })
df_full
```

```python
df_growth_daily = df_full[1:len(df_full)]
df_growth_daily = df_growth_daily.reset_index(drop=True)

df_growth_daily['Btc/D, %'] = round(((df_growth_daily['Btc close'] - df_growth_daily['Btc open']) / ((df_growth_daily['Btc close'] + df_growth_daily['Btc open']) / 2) * 100 ), 2)
df_growth_daily['Ltc/D, %'] = round(((df_growth_daily['Ltc close'] - df_growth_daily['Ltc open']) / ((df_growth_daily['Ltc close'] + df_growth_daily['Ltc open']) / 2) * 100 ), 2)
df_growth_daily['Btc vol/D, %'] = round(((df_growth_daily['Btc volume'] - df_full['Btc volume']) / ((df_growth_daily['Btc volume'] + df_full['Btc volume']) / 2) * 100 ), 2)
df_growth_daily['Ltc vol/D, %'] = round(((df_growth_daily['Ltc volume'] - df_full['Ltc volume']) / ((df_growth_daily['Ltc volume'] + df_full['Ltc volume']) / 2) * 100 ), 2)
df_growth_daily['Btc adr/D, %'] = round(((df_growth_daily['Btc adress'] - df_full['Btc adress']) / ((df_growth_daily['Btc adress'] + df_full['Btc adress']) / 2) * 100 ), 2)
df_growth_daily['Ltc adr/D, %'] = round(((df_growth_daily['Ltc adress'] - df_full['Ltc adress']) / ((df_growth_daily['Ltc adress'] + df_full['Ltc adress']) / 2) * 100 ), 2)
df_growth_daily['Btc tx/D, %'] = round(((df_growth_daily["Btc tx's"] - df_full["Btc tx's"]) / ((df_growth_daily["Btc tx's"] + df_full["Btc tx's"]) / 2) * 100 ), 2)
df_growth_daily['Ltc tx/D, %'] = round(((df_growth_daily["Ltc tx's"] - df_full["Ltc tx's"]) / ((df_growth_daily["Ltc tx's"] + df_full["Ltc tx's"]) / 2) * 100 ), 2)
df_growth_daily['Btc fees/D, %'] = round(((df_growth_daily['Btc fees'] - df_full['Btc fees']) / ((df_growth_daily['Btc fees'] + df_full['Btc fees']) / 2) * 100 ), 2)
df_growth_daily['Ltc fees/D, %'] = round(((df_growth_daily['Ltc fees'] - df_full['Ltc fees']) / ((df_growth_daily['Ltc fees'] + df_full['Ltc fees']) / 2) * 100 ), 2)

df_growth_daily = df_growth_daily.drop(columns = ['Btc close', 'Btc open', 'Ltc close', 'Ltc open',
                                      'Btc volume', 'Ltc volume', 'Btc adress', 'Ltc adress',
                                      "Btc tx's", "Ltc tx's", 'Btc fees', 'Ltc fees'])
df_growth_daily
```

```python
df_growth_monthly = df_growth_daily.set_index('Date').resample('M')[['Btc/D, %', 'Ltc/D, %','Btc vol/D, %',
                                                                     'Ltc vol/D, %','Btc adr/D, %', 'Ltc adr/D, %',
                                                                     'Btc tx/D, %', 'Ltc tx/D, %','Btc fees/D, %',
                                                                     'Ltc fees/D, %']].sum()
df_growth_monthly.reset_index(level=0, inplace=True)
df_growth_monthly.rename(columns = {'Btc/D, %' : 'Btc/M, %', 'Ltc/D, %' : 'Ltc/M, %',
                     'Btc vol/D, %' : 'Btc vol/M, %', 'Ltc vol/D, %' : 'Ltc vol/M, %',
                     'Ltc adr/D, %' : 'Ltc adr/M, %', 'Btc tx/D, %' : 'Btc tx/M, %',
                     'Ltc tx/D, %' : 'Ltc tx/M, %', 'Btc fees/D, %' : 'Btc fees/M, %',
                     'Ltc fees/D, %' : 'Ltc fees/M, %', 'Btc adr/D, %': 'Btc adr/M, %'},inplace = True) 
df_growth_monthly                                                           
```

```python
df_ltc_btc = df_full[['Date', 'Btc close', 'Btc volume', 'Btc adress', "Btc tx's",
       'Btc fees', 'Ltc close', 'Ltc volume', 'Ltc adress',"Ltc tx's", 'Ltc fees']]
df_ltc_btc = df_ltc_btc.reset_index(drop=True)
df_ltc_btc['Ltc/btc'] = df_full['Ltc close'] / df_full['Btc close']
df_ltc_btc['Ltc/btc volume'] = df_full['Ltc volume'] / df_full['Btc volume']
df_ltc_btc['Ltc/btc fees'] = df_full['Ltc fees'] / df_full['Btc fees']
df_ltc_btc["Ltc/btc tx's"] = df_full["Ltc tx's"] / df_full["Btc tx's"]
df_ltc_btc["Ltc/btc adress"] = df_full["Ltc adress"] / df_full["Btc adress"]
df_ltc_btc = df_ltc_btc.drop(columns= ['Btc close', 'Btc volume', 'Btc adress', "Btc tx's",
       'Btc fees', 'Ltc close', 'Ltc volume', 'Ltc adress',"Ltc tx's", 'Ltc fees'])
df_ltc_btc
```

```python
import plotly.express as px

fig = px.line(df_full, x='Date', y='Btc close')
fig.add_vrect(x0="2012-11-28", x1="2016-07-09",annotation_text="Halving #1",
              fillcolor="green", opacity=0.1, line_width=0)
fig.add_vrect(x0="2016-07-09", x1="2020-05-11",annotation_text="Halving #2",
              fillcolor="blue", opacity=0.1, line_width=0)
fig.add_vrect(x0="2020-05-11", x1="2024-05-05", annotation_text="Halving #3",
              fillcolor="red", opacity=0.1, line_width=0)
fig.update_annotations(align= 'center')
fig.update_layout(title="Bitcoin chart + Halvings",
                  xaxis_title="Date", title_x= 0.5,
                  yaxis_title="Price USD",
                 margin=dict(l=0, r=0, t=50, b=0))
fig.show()

fig = px.line(df_full, x='Date', y='Ltc close')
fig.add_vrect(x0="2015-08-25", x1="2019-08-05", annotation_text="Halving #1",annotation_position="bottom left",
              fillcolor="green", opacity=0.2, line_width=0)
fig.add_vrect(x0="2019-08-05", x1="2023-08-04",annotation_text="Halving #2",annotation_position="bottom left",
              fillcolor="blue", opacity=0.2, line_width=0)
fig.update_layout(title="Litecoin chart + Halvings",
                  xaxis_title="Date", title_x= 0.5,
                  yaxis_title="Price USD",
                 margin=dict(l=0, r=0, t=50, b=0))
fig.show()

fig = px.line(df_ltc_btc, x='Date', y='Ltc/btc')
fig.add_vrect(x0="2015-08-25", x1="2019-08-05",annotation_text="Ltc Halving #1",annotation_position="bottom left",
              fillcolor="green", opacity=0.2, line_width=0)
fig.add_vrect(x0="2019-08-05", x1="2023-08-04",annotation_text="Ltc Halving #2",annotation_position="bottom left",
              fillcolor="blue", opacity=0.2, line_width=0)
fig.add_vrect(x0="2012-11-28", x1="2016-07-09",annotation_text="BTC Halving #1",
              fillcolor="green", opacity=0.1, line_width=0)
fig.add_vrect(x0="2016-07-09", x1="2020-05-11",annotation_text="BTC Halving #2",
              fillcolor="blue", opacity=0.1, line_width=0)
fig.add_vrect(x0="2020-05-11", x1="2024-05-05",annotation_text="BTC Halving #3",
              fillcolor="red", opacity=0.1, line_width=0)
fig.update_layout(title="Litecoin/Bitcoin chart + Halvings each other",
                  xaxis_title="Date", title_x= 0.5,
                  yaxis_title="Price USD",
                 margin=dict(l=0, r=0, t=50, b=0))
fig.show()
```

Можно сделать вывод о том, что:
    - до форка лайткоин растет к биткоину в последней четверти времени
    - во время форка и после на протяжении половины времени он снижается к биткоину

```python
df_monthly = df_ltc_btc.set_index('Date').resample('M')[['Ltc/btc volume', 'Ltc/btc fees', 'Ltc/btc adress', "Ltc/btc tx's"]].sum()
df_monthly.reset_index(level=0, inplace=True)

import plotly.graph_objects as go
fig = go.Figure()
fig.add_trace(
    go.Bar(
        x=df_monthly['Date'],
        y=df_monthly["Ltc/btc tx's"]))
fig.update_layout(title="Litecoin / Bitcoin transaction count",
                  xaxis_title="Date", title_x= 0.5,
                  yaxis_title="Ratio",
                 margin=dict(l=0, r=0, t=30, b=0))
fig.show()

fig = go.Figure()
fig.add_trace(
    go.Bar(
        x=df_monthly['Date'],
        y=df_monthly["Ltc/btc adress"]))
fig.update_layout(title="Litecoin / Bitcoin adresses count",
                  xaxis_title="Date", title_x= 0.5,
                  yaxis_title="Ratio",
                 margin=dict(l=0, r=0, t=30, b=0))
fig.show()

fig = go.Figure()
fig.add_trace(
    go.Bar(
        x=df_growth_monthly['Date'],
        y=df_growth_monthly['Btc adr/M, %'], name="Btc adresses monthly, %"))
fig.add_trace(
    go.Bar(
        x=df_growth_monthly['Date'],
        y=df_growth_monthly['Ltc adr/M, %'], name="Ltc adresses monthly, %"))
fig.update_layout(title="Percentage increase Litecoin and Bitcoin adresses monthly, %",
                  xaxis_title="Date", title_x= 0.5,
                  yaxis_title="Ratio",
                 margin=dict(l=0, r=0, t=30, b=0))
fig.update_layout(legend=dict(
    yanchor="top",
    y=0.98,
    xanchor="left",
    x=0.01
))
fig.show()


fig = go.Figure()
fig.add_trace(
    go.Bar(
        x=df_monthly['Date'],
        y=df_monthly['Ltc/btc volume'], name="Btc/Usd monthly, %"))
fig.update_layout(title="Litecoin / Bitcoin transaction volume",
                  xaxis_title="Date", title_x= 0.5,
                  yaxis_title="Ratio",
                 margin=dict(l=0, r=0, t=30, b=0))
fig.show()
```

На первом и втором графике видно, что идет стабильный рост адресов и транзакций в сети лайткоин в сравнении с биткоином
На третьем начиная с мая 2017 года прирост адресов лайткоина больше, нежели у биткоина
Однако на 4 графике мы видим, что обьемы транзакций июня 2019 года снижаются
Все это нам говорит о том, что лайткоин используют для ежедневных трат, благодаря его преимуществам над биткоином.
Однако биткоин используют в основном для накопления, а не для покупки чашечки кофе.
